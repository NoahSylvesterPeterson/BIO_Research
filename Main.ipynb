```python
from BioInfer import *
from LZCompression import Mutual_Compression_ratio, Mutual_Compression_ratio_Crossed,  Mutual_Compression_ratio_Conditional
```

```python
seqs = ReadMultiFromFASTA("alpha_tubulin_short_names.fasta", Discriptor = True)
print(seqs)

f =  open("Alpha_Tubulin_ClustalMatrix.txt", "r")
lines = f.readlines()
f.close()

CPI_Matrix = []
order =[]
for line in lines[6:]:
    l = line.split()
    CPI_Matrix.append(l[2:])
    order.append(l[1])

print(CPI_Matrix)
print(order)
```

```python
def MatrixLine(fileName, order, MuteInfo,seqs):
    f = open(fileName,"w")
    for seq1 in order:
        line = [seq1]
        for seq2 in order:
            line.append(str(MuteInfo(seqs[seq1],seqs[seq2])))
        f.write(",".join(line))
    f.close()

def Column(fileName, order, seqs, Custal_Matix):
    f = open(fileName,"w")
    f.write("Definition1,Definition2,Definition3,Definition4,Clustal PI \n")
    for i in range(len(order)):
        for j in range(len(order)):
            line = []
            line.append(str(Mutual_Compression_ratio(seqs[order[i]],seqs[order[j]])))
            line.append(str(Mutual_Compression_ratio(seqs[order[i]],seqs[order[j]], zip = True)))
            line.append(str(Mutual_Compression_ratio_Crossed(seqs[order[i]],seqs[order[j]])))
            line.append(str(Mutual_Compression_ratio_Conditional(seqs[order[i]],seqs[order[j]])))
            line.append(Custal_Matix[i][j])
            f.write(",".join(line)+"\n")
    f.close()
```

```python
Column("oupt_advanced.csv", order, seqs, CPI_Matrix)
```
